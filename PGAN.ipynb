```python
import torch
import matplotlib.pyplot as plt
import torchvision
```

```python
use_gpu = True if torch.cuda.is_available() else False
total_images = 8

model = torch.hub.load('facebookresearch/pytorch_GAN_zoo:hub','PGAN', model_name='celebAHQ-512', pretrained=True, useGPU=use_gpu)
```

```python
noise, _ = model.buildNoiseData(total_images)

with torch.no_grad():
    images = model.test(noise)
```

```python
plt.imshow(torchvision.utils.make_grid(images.clamp(min=-1, max=1), scale_each=True, normalize=True).permute(1, 2, 0).cpu().numpy())
plt.axis('off')
plt.show()
```
